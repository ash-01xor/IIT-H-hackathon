### Building Model:

```python
from sklearn.tree import DecisionTreeRegressor
import xgboost as xgb
from lightgbm import LGBMRegressor
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
```

```python
X_train = train[train['Year'] < 2021].drop(['Power Level'], axis=1)
y_train = train[train['Year'] < 2021]['Power Level']

X_test = test[test['Year'] == 2022].drop(['Power Level'], axis=1)
y_test = test[test['Year'] == 2022]['Power Level']
```

```python
params={'random_state': 33,'n_estimators':5000,
 'min_data_per_group': 5,
 'boosting_type': 'gbdt',
 'num_leaves': 256,
 'max_dept': -1,
 'learning_rate': 0.02,
 'subsample_for_bin': 200000,
 'lambda_l1': 1.074622455507616e-05,
 'lambda_l2': 2.0521330798729704e-06,
 'n_jobs': -1,
 'cat_smooth': 1.0,
 'silent': True,
 'importance_type': 'split',
 'metric': 'rmse',
 'feature_pre_filter': False,
 'bagging_fraction': 0.8206341150202605,
 'min_data_in_leaf': 100,
 'min_sum_hessian_in_leaf': 0.001,
 'bagging_freq': 6,
 'feature_fraction': 0.5,
 'min_gain_to_split': 0.0,
 'min_child_samples': 20}
```

```python
kfold = KFold(n_splits=5, shuffle=True, random_state=42)
mse_values = []

for fold, (train_idx, val_idx) in enumerate(kfold.split(X_train, y_train)):
    
    X_train_fold, y_train_fold = X_train.iloc[train_idx], y_train.iloc[train_idx]
    X_val_fold, y_val_fold = X_train.iloc[val_idx], y_train.iloc[val_idx]
    
    model =  model = LGBMRegressor(**params)
    model.fit(X_train_fold, y_train_fold)
    
    y_pred_fold = model.predict(X_val_fold)
    mse_fold = mean_squared_error(y_val_fold, y_pred_fold)
    mse_values.append(mse_fold)
    
    print('Fold %d, MSE: %.4f' % (fold+1, mse_fold))

avg_mse = np.mean(mse_values)
print('Average MSE across all folds: %.4f' % avg_mse)
```

```python
model = LGBMRegressor(**params)
model.fit(X_train, y_train)

y_pred_test = model.predict(X_test)

test_mse = mean_squared_error(y_test, y_pred_test)
print('Test MSE: %.4f' % test_mse)
```
